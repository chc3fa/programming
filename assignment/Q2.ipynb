```python
import numpy as np
import matplotlib.pyplot as plt
#Step 1-4
x = np.linspace(0, 1, num=50)
y = np.log(x)
z = np.exp(x)
plt.scatter(x, y, label='Natural Log')
plt.scatter(x, z, label='Exponential')

#Step 6 and 7
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Natural Log and Exponential Functions")
plt.legend(loc = 'lower right')
plt.show()

#Step 8-12
x2 = np.linspace(-6.5, 6.5, num=100)
y2 = np.sin(x2)
z2 = np.cos(x2)
plt.scatter(x2, y2)
plt.scatter(x2, z2)
plt.show()
#Step 11 and 12
plt.plot(x2, y2, label ='Sin')
plt.plot(x2, z2, label='Cosine')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sin and Cosine from -6.5 to 6.5")
plt.legend(loc = 'lower left')
plt.show()
```
